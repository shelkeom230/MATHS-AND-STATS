```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("usincome.csv",index_col=None,names=['income','count'],skiprows=1)
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
```

```python
# transform x axis on a scale of log 

sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale='log')
```

```python
# using matplotlib 
plt.figure(figsize=(11.7, 8.27))  # Set figure size
plt.bar(df['income'].astype(str), df['count'])  # Create bar plot (use barh for horizontal bar plot)
plt.xticks(rotation=45, ha='right',color='purple')  # Rotate x-axis labels
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Income vs Count')
plt.show()
```

```python

plt.figure(figsize=(11.7, 8.27))  # Set figure size
plt.bar(df['income'].astype(str), df['count'])  # Create bar plot
plt.xscale('log')  # Apply logarithmic scale to x-axis
plt.xticks(rotation=45, ha='right',fontsize=15,color='red')  # Rotate x-axis labels
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Income vs Count (Log Scale)')
plt.show()
```
