```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("movie_revenues.csv")
df.head()
```

```python
df['revenue_mln']=df['revenue'].apply(lambda x:x/1000000)
```

```python
df.revenue_mln.describe()
```

```python
df.head(2)
```

```python
# get the regular z score 
df.revenue_mln.describe()
```

```python
_,mean,std,*_ = df.revenue_mln.describe()
mean,std
```

```python
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
df['zscore']=df.revenue_mln.apply(lambda x:get_z_score(x,mean,std))
```

```python
df.head(3)
```

```python
df[df.zscore>3]
```

```python
# get the MAD 
def get_mad(s):
    median=np.median(s)
    diff=abs(s-median)
    MAD=np.median(diff)
    return MAD
```

```python
MAD=get_mad(df.revenue_mln)
median=np.median(df.revenue_mln)
MAD , median
```

```python
# get modified z score 
def get_modified_zscore(x,median,MAD):
    return 0.6745*(x-median)/MAD 
```

```python
get_modified_zscore(2787,median,MAD)
```

```python
get_modified_zscore(1513,median,MAD)
```

```python
df['mod_zscore']=df.revenue_mln.apply(lambda x:get_modified_zscore(x,median,MAD))
```

```python
df.head(3)
```

```python
df[df.mod_zscore>3]
```
